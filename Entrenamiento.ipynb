```
import sys
import os
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import optimizers
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dropout, Flatten, Dense, Activation
from tensorflow.python.keras.layers import Convolution2D, MaxPooling2D
from tensorflow.python.keras import backend as K
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
K.clear_session()

train_data = '/content/drive/MyDrive/Avance2-ML/train_data'
validation_data = '/content/drive/MyDrive/Avance2-ML/validation_data'
```

```
##Parametros

epochs=50
height, width = 150, 150
batch_size = 8
steps = 120
validation_steps = 256
conv_filters_1 = 32
conv_filters_2 = 64
filter_size_1 = (3,3)
filter_size_2 = (2,2)
pool_size = (2,2)
classes = 3 ##Clases a clasificar
lr = 0.0004
```

```
##Image pre-processing

training_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)

validation_datagen = ImageDataGenerator(
    rescale = 1./255
)

training_image = training_datagen.flow_from_directory(
    train_data,
    target_size = (height,width),
    batch_size = batch_size,
    class_mode = 'categorical'
)

validation_image = validation_datagen.flow_from_directory(
    validation_data,
    target_size = (height,width),
    batch_size = batch_size,
    class_mode = 'categorical'
)
```

```
## Creating neuronal network

cnn=Sequential()

cnn.add(Convolution2D(conv_filters_1, filter_size_1, padding='same', input_shape=(height,width,3), activation='relu'))
cnn.add(MaxPooling2D(pool_size=pool_size))

cnn.add(Convolution2D(conv_filters_2, filter_size_2, padding='same', activation='relu'))
cnn.add(MaxPooling2D(pool_size=pool_size))

cnn.add(Flatten())
cnn.add(Dense(256,activation='relu'))
cnn.add(Dropout(0.5))
cnn.add(Dense(classes,activation='softmax'))

cnn.compile(loss='categorical_crossentropy',
            optimizer=optimizers.Adam(learning_rate=lr),
            metrics=['accuracy'])
```

```
##Entrenar

cnn.fit_generator(
    training_image,
    steps_per_epoch=steps,
    epochs=epochs,
    validation_data=validation_image,
    validation_steps=validation_steps)

target_dir='/content/drive/MyDrive/Avance2-ML/model/'

if not os.path.exists(target_dir):
    os.mkdir(target_dir)
cnn.save('/content/drive/MyDrive/Avance2-ML/model/model.h5')
cnn.save_weights('/content/drive/MyDrive/Avance2-ML/model/weight.h5')
```
